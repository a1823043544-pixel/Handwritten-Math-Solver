```python
import tensorflow as tf
import os

# 这里填你训练好的模型文件名
model_path = 'math_model.h5'  # 或者 'math_solver_model.h5'

if os.path.exists(model_path):
    print(f"正在加载模型: {model_path} ...")
    try:
        model = tf.keras.models.load_model(model_path)
        
        # 这行代码会打印出完整的架构表
        model.summary()
        
    except Exception as e:
        print(f"加载失败: {e}")
else:
    print(f"找不到文件: {model_path}，请检查文件名是否正确。")
```
